```python
x=data.iloc[:,:12].values
y=data.iloc[:,12].values
```

```python
kf=KFold(n_splits=5)
```

```python
list(kf.split(data))
```

```python
 model=LinearRegression()
```

```python
RMSE=[]
for i in list(kf.split(data)):
  model.fit(x[i[0],:],y[i[0]])
  y_pred=model.predict(x[i[1],:])
  RMSE.append(np.sqrt(mean_squared_error(y_pred,y[i[1]])))
  RMSE
```

```python
np.mean(RMSE)
```
